```python
# Read Consolidated File
csvpath_consolidated_file= "CONSOLIDATED_CONSUMER_SPENDING.csv"
consolidated_file = pd.read_csv(csvpath_consolidated_file)
```

```python
#Show intial DF
consolidated_file.head()
```

```python
#identify data types

consolidated_file.dtypes
```

```python
# Remove commas & convert from object to numeric

columns = ["$", "$ YA", "Units", "Units YA", "Any Promo $", "Any Promo $ YA", "No Promo $", "No Promo $ YA"]

for col in columns:
    consolidated_file[col]=consolidated_file[col].replace(',','',regex=True).apply(pd.to_numeric, errors='coerce')
```

```python
#Fill NA

columns = ["$", "$ YA", "Units", "Units YA", "Any Promo $", "Any Promo $ YA", "No Promo $", "No Promo $ YA"]

for col in columns:
    consolidated_file[col]=consolidated_file[col].fillna(0)
```

```python
# Confirm that they have converted to objects

consolidated_file.dtypes
```

```python
#Drop irrelevant & innacurate columns

consolidated_file_two = consolidated_file.drop(columns=["TTL","Base % Chg."])
```

```python
# Rename columns for easier readability
consolidated_file_three = consolidated_file_two.rename(columns={"SMM":"State","YR": "Year",
                                                     "BCD":"Category", "BCSC":"Segment",
                                                     "Any Promo $":"Promotional Dollars", 
                                                                 "Any Promo $ YA":"Promotional Dollars Year Ago",
                                                    "No Promo $":"Non Promoted Dollars", 
                                                                 "No Promo $ YA":"Non Promoted Dollars Year Ago"})
```

```python
#Reindex

consolidated_file_four = consolidated_file_three.set_index(["State"])
```

```python
consolidated_file_four.head()
```

```python
#Calculations 

Dollar_Change = consolidated_file_four["$"]-consolidated_file_four["$ YA"]
Dollar_Percent_Change = ((consolidated_file_four["$"]/consolidated_file_four["$ YA"])-1)*100
Unit_Change = consolidated_file_four["Units"]-consolidated_file_four["Units YA"]
Unit_Percent_Change = ((consolidated_file_four["Units"]/consolidated_file_four["Units YA"])-1)*100
Trade_Up_Index = Dollar_Change/Unit_Change
Percent_On_Promo = consolidated_file_four["Promotional Dollars"]/consolidated_file_four["$"]
Percent_On_Promo_YA = consolidated_file_four["Promotional Dollars Year Ago"]/consolidated_file_four["$ YA"]
Share_Promo_Growing = ((Percent_On_Promo/Percent_On_Promo_YA)-1)*100
Promo_Percent_Change = ((consolidated_file_four["Promotional Dollars"]/consolidated_file_four["Promotional Dollars Year Ago"])-1)*100
Market_Dilution = Dollar_Percent_Change - Promo_Percent_Change
```

```python
#Add calculations to dataframe

consolidated_file_four["$ Abs. Chg."] = Dollar_Change
consolidated_file_four["$ % Chg."] = Dollar_Percent_Change
consolidated_file_four["Unit Abs. Chg."] = Unit_Change
consolidated_file_four["Unit % Chg."] = Unit_Percent_Change 
consolidated_file_four["Trade Up Index"] = Trade_Up_Index 
consolidated_file_four["Percent On Promo"] = Percent_On_Promo
consolidated_file_four["Share of Promo Growth"] = Share_Promo_Growing
consolidated_file_four["Promo % Chg."] = Promo_Percent_Change
consolidated_file_four["Market Dilution"] = Market_Dilution 
```

```python
#Save Dataframe in CSV

consolidated_file_four.to_csv("CLEANSED FILE.csv", index=False, header=True)
```
